# 1. In Python / LUA, read five arbitrary 128x128 pixel color images, display and print the contents of arrays (matrices, tensors) containing RGB channels.

```
# Import libraries 
from PIL import Image
import requests
from io import BytesIO
import numpy as np
```

```
# Add 128x128px pictures with urls
url = ["https://icons.iconarchive.com/icons/aha-soft/free-global-security/128/CCTV-Camera-icon.png", 
       "https://icons.iconarchive.com/icons/gakuseisean/aire/128/Images-icon.png",
       "https://icons.iconarchive.com/icons/yohproject/cute/128/camera-2-icon.png",
       "https://icons.iconarchive.com/icons/treetog/junior/128/folder-blue-pictures-icon.png",
       "https://icons.iconarchive.com/icons/chromatix/aerial/128/image-icon.png"]
```

```
# Read in the images, display them and print out the RGB values
rgba = []
for i in range(len(url)):
  response = requests.get(url[i])
  img = Image.open(BytesIO(response.content))
  display(img)
  rgba.append(np.array(img))
  print(rgba[i])
```

# 2. Calculate the average and standard deviation of each pixel per channel for the five images, and then convert them to a dataset with an expected value of 0 and a standard deviation of 1. Then check the expected value and standard deviation of the resulting dataset.

```
# Initialize the lists for average and variance
avg = []
var = []

# Loop over the colors (red, green, blue - alpha is not important in this exercise)
for color in range(3):
  for im in range(len(rgba)):
    # Append mean and variance for each color in the pictures
    avg.append(np.mean(rgba[im][:,:,color]))
    var.append(np.var(rgba[im][:,:,color]))

    # Print out the values - ('im+1' for labeling the pictures from 1-5 not 0-4)
    if color == 0:
      print("Average of red is {0:.2f} and variance is {1:.2f} in picture {2}".format(avg[color*im + im], var[color*im + im], im+1))
    elif color == 1:
      print("Average of green is {0:.2f} and variance is {1:.2f} in picture {2}".format(avg[color*im + im], var[color*im + im], im+1))
    else:
      print("Average of blue is {0:.2f} and variance is {1:.2f} in picture {2}".format(avg[color*im + im], var[color*im + im], im+1))
```

```
# Import for value transformation
from sklearn.preprocessing import PowerTransformer
# Import for ignoring runtime warnings in the following cells during calculating variance
import warnings
warnings.filterwarnings('ignore')
```

```
# Power transform method to make data Gaussian-like
pt = PowerTransformer()
```

```
# Initialize the lists for the new channel values and for average and variance
rgba_new = []
avg_new = []
var_new = []

# Loop over the colors (red, green, blue - alpha is not important in this exercise)
for color in range(3):
  for im in range(len(rgba)):
    # Fit and tranform the each color in the pictures
    rgba_new.append(pt.fit_transform(rgba[im][:,:,color]))

    # Append the new mean and variance 
    avg_new.append(np.mean(rgba_new[color*im + im]))
    var_new.append(np.var(rgba_new[color*im + im]))

    # Print out the new values - ('im+1' for labeling the pictures from 1-5 not 0-4)
    if color == 0:
      print("Average of red is {0:.2f} and variance is {1:.2f} in picture {2}".format(avg_new[color*im + im], var_new[color*im + im], im+1))
    elif color == 1:
      print("Average of green is {0:.2f} and variance is {1:.2f} in picture {2}".format(avg_new[color*im + im], var_new[color*im + im], im+1))
    else:
      print("Average of blue is {0:.2f} and variance is {1:.2f} in picture {2}".format(avg_new[color*im + im], var_new[color*im + im], im+1));
```

# 3. Read any audio file in Python / LUA and display its spectrogram.

```
# Import libraries 
import matplotlib.pyplot as plt
from scipy import signal
from scipy.io import wavfile
```

```
# Read sample rate and data from the wav file
sample_rate, data = wavfile.read('StarWars60.wav')

# Compute the spectrogram of wav with consecutive Fourier transforms
frequencies, segment_times, data_spectrogram = signal.spectrogram(data, sample_rate)

# Create plot
plt.figure(figsize=(20,10))
plt.pcolormesh(segment_times, frequencies, data_spectrogram, shading='auto')
# Add colorbar
plt.colorbar()
plt.ylabel('Frequency [Hz]')
plt.ylim(top=3000)
plt.xlabel('Time [s]')
plt.show()
```

# 4. Using a Python / LUA script, download the textual content of https://en.wikipedia.org/wiki/Deep_learning, display the textual content, and display the frequency of the letters in the content in a histogram.

```
# Import library
import urllib.request
from urllib.request import urlopen
from bs4 import BeautifulSoup
import string
import matplotlib.pyplot as plt
```

```
# Open url and show html context
url = "https://en.wikipedia.org/wiki/Deep_learning"
html = urlopen(url).read()
```

```
# Getting text from html in this cell is based on https://stackoverflow.com/a/24618186
# Pulling data out of html with BeautifulSoup
soup = BeautifulSoup(html, features="html.parser")

# Remove script and style elements
for script in soup(["script", "style"]):
    script.extract() 

# Get text function
deeplearning_text = soup.get_text()

# Get lines and remove white-space characters
lines = (line.strip() for line in deeplearning_text.splitlines())

# Get multilines into line
line_elements = (phrase.strip() for line in lines for phrase in line.split("  "))

# Join line elements together
deeplearning_text = '\n'.join(line_element for line_element in line_elements if line_element)
```

```
# Print out the text
print(deeplearning_text)
```

```
# Get the abc from string library
abc = list(string.ascii_lowercase) 
```

```
# Get the occurrence of each letter
nums = np.array([])
for letter in abc:
  nums = np.append(nums, deeplearning_text.count(letter))

# Create a range with the number of letter
n=np.arange(len(nums))
```

```
# Frequency plot
plt.bar(n, nums)
plt.title('Number of occurrences for each letter in the text')
plt.xticks(n, abc);
```

```
# Percent plot
plt.bar(n, nums/sum(nums))
plt.title('Percentage of occurrences for each letter in the text')
plt.xticks(n, abc);
```
